# Fake News Classifier

```python
import pandas as pd

df = pd.read_csv('Datasets/train.csv')
```

```python
df.head()
```

```python
X = df.drop('label',axis=1)
```

```python
X.head()
```

```python
df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.head(10)
```

```python
y = df.label
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
import re
corpus = []
for i in range(0,len(df)):
    rev = re.sub("[^a-zA-Z]",' ',df['title'][i])
    rev = rev.lower()
    rev = rev.split()

    rev = [ps.stem(word) for word in rev if not word in stopwords.words('english')]
    rev = ' '.join(rev)
    corpus.append(rev)
```

```python
corpus
```

```python
#Using Bag of words

cv = CountVectorizer(max_features=5000, ngram_range=(1,3))
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.35,random_state=952)
```

```python
import itertools
import matplotlib.pyplot as plt
import numpy as np

def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
```

```python
nb.fit(X_train,y_train)
```

# Predictions, Accuracy and Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score

pred = nb.predict(X_test)
acc = accuracy_score(y_test, pred)
cm = confusion_matrix(y_test, pred)
plot_confusion_matrix(cm, classes=['FAKE', 'REAL'])
```

```python
print(f"Accuracy = {acc}")
```

# Passive aggresive Classifier

```python
from sklearn.linear_model import PassiveAggressiveClassifier
```

```python
cl = PassiveAggressiveClassifier()
cl.fit(X_train,y_train)
```

```python
pred2 = cl.predict(X_test)
acc2 = accuracy_score(y_test, pred2)
cm2 = confusion_matrix(y_test, pred2)
plot_confusion_matrix(cm2, classes=['FAKE', 'REAL'])
```

```python
print(f"Accuracy = {acc2}")
```

# USING LSTM

## One Hot Representation

```python
data = pd.read_csv('Datasets/train.csv')
```

```python
messages = data.drop('label',axis=1)
```

```python
messages.dropna(inplace=True)
```

```python
messages.reset_index(inplace=True)
```

```python
messages
```

# Data Pre-Processing

```python
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import re
ps = PorterStemmer()
```

```python
corpus2 = []

for i in range (len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    review = " ".join(review)
    corpus2.append(review)
```

# One Hot Representation

```python
from keras.layers import Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.preprocessing.text import one_hot
from keras.layers import LSTM
from keras.layers import Dense
```

```python
voc_size=5000
```

```python
one_hot_rep = [one_hot(words,voc_size) for words in corpus2]
one_hot_rep
```

# Embedding Representation

```python
sent_length = 10
embedded_docs = pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
embedded_docs
```

# Creating Model

```python
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=952)
```

# Model Training

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

# Performance Metrics And Accuracy

```python
y_pred=(model.predict(X_test) > 0.5).astype("int32")
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm3=confusion_matrix(y_test,y_pred)
plot_confusion_matrix(cm3, classes=['FAKE', 'REAL'])
```

```python
accuracy_score(y_test,y_pred)
```
